```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pickle

pca = PCA(n_components=2)
```

```python
f = open('output_vectors.pckl', 'rb')
output_vectors = pickle.load(f)
f.close()

f = open('output_actuators.pckl', 'rb')
output_actuators = pickle.load(f)
f.close()
```

```python
pca.fit(output_vectors,output_actuators)
X_embedded = pca.fit_transform(output_vectors,output_actuators)

fig,ax = plt.subplots(1)
ax.scatter(X_embedded[:,0], X_embedded[:,1], c=output_actuators, s=5)
# ax.set_yticks([])
# ax.set_yticklabels([])
# ax.set_xticks([])
# ax.set_xticklabels([])
# ax.axis([-6000,20000,-1000,1000])
fig.dpi=100
# fig.legend(loc="upper left")
# plt.savefig('/Users/anirudhnatarajan/Documents/file.eps')
# plt.savefig('/Users/anirudhnatarajan/Documents/file.png')

# purple = 0, blue = 1, green = 2, yellow = 3
# Method 3
```
